$$r_k^{(n)} = \frac{\pi_k \frac{1}{\prod_{i=1}^{I} \sqrt{2\pi}\sigma_i^{(k)}} \exp \left( -\sum_{i=1}^{I} (m_i^{(k)} - x_i^{(n)})^2 / 2(\sigma_i^{(k)})^2 \right)}{\sum_{k'} (\text{numerator, with } k' \text{ in place of } k)}$$

$$m_i^{(k)} = \frac{\sum_{n} r_k^{(n)} x_i^{(n)}}{R^{(k)}}$$

$$\Sigma^{(k)} = \frac{\sum_{n} r_k^{(n)}(\mathbf{x}^{(n)} - \mathbf{m}^{(k)})(\mathbf{x}^{(n)} - \mathbf{m}^{(k)})^T}{R^{(k)}}$$

$$Will\ scale\ Gaussians\ only\ horizontaly\ and\ vertically$$
$$\sigma_i^{2(k)} = \frac{\sum_{n} r_k^{(n)}(x_i^{(n)} - m_i^{(k)})^2}{R^{(k)}}$$

$r_k^{(n)}$: Responsibility (membership weight) \
$\pi_k$: Mixing coefficient (prior probability) for cluster $k$ \
$m_i^{(k)}$: Mean of cluster $k$ in dimension $i$ \
$\sigma_i^{(k)}$: Standard deviation of cluster $k$ in dimension $i$ \
$x_i^{(n)}$: Value of the $n$-th data point in dimension $i$ \
$R^{(k)}$: Total responsibility assigned to cluster $k$ ($\sum_n r_k^{(n)}$)
