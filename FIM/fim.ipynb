### Setup: The Key Facts

#### 1. Definition of FIM 

$$F(\theta) = \mathbb{E}_{X \sim p_\theta}\left[\nabla_\theta \log p(X \mid \theta) \nabla_\theta \log p(X \mid \theta)^\top\right] = -\mathbb{E}_{X \sim p_\theta}\left[\nabla_\theta^2 \log p(X \mid \theta)\right]$$

#### 2. Local behavior of the log-likelihood near the MLE: $\hat{\theta}$

$$\ell(\theta) = \sum_{i=1}^{n} \log p(x_i \mid \theta)$$
$$\nabla \ell(\theta) = \sum_{i} \nabla_\theta \log p(x_i \mid \theta)$$
$$\nabla^2 \ell(\theta) = \sum_{i} \nabla_\theta^2 \log p(x_i \mid \theta)$$
$$\ell(\hat{\theta} + \delta) \approx \ell(\hat{\theta}) + \underbrace{\nabla \ell(\hat{\theta})^\top \delta}_{= 0 \text{ at MLE}} + \frac{1}{2} \delta^\top \nabla^2 \ell(\hat{\theta}) \delta$$
$$J(\theta) = -\nabla_\theta^2 \ell(\theta) = -\sum_{i=1}^{n} \nabla_\theta^2 \log p(x_i \mid \theta)$$
$$\ell(\hat{\theta} + \delta) - \ell(\hat{\theta}) \approx -\frac{1}{2} \delta^\top J(\hat{\theta}) \delta$$

#### 3. Connection between observed and expected FIM
$$\frac{1}{n} J(\theta) \xrightarrow{n \to \infty} F(\theta)$$

#### 4. FIM as the local curvature of KL divergence:

$$D_{KL}(p_\theta \| p_{\theta + \delta}) \approx \frac{1}{2} \delta^\top F(\theta) \delta$$

## Forward Direction: Sensitive Distribution ⇒ Informative Data

**Premise.** A small change $\delta$ in $\theta$ produces a very different distribution — formally, $D_{KL}\!\left(p_\theta \,\|\, p_{\theta + \delta}\right)$ is large even for small $\|\delta\|$.

**Step 1 — By (4):** $D_{KL} \approx \tfrac{1}{2}\, \delta^\top F(\theta)\, \delta$. Since the left side is large for small $\delta$, the matrix $F(\theta)$ must be large.

**Step 2 — By (3):** For $n$ i.i.d. observations, the observed Fisher information $J(\theta) \approx n\, F(\theta)$ is also large.

**Step 3 — By (2):** Near the MLE, $\ell(\hat\theta + \delta) - \ell(\hat\theta) \approx -\tfrac{1}{2}\, \delta^\top J(\hat\theta)\, \delta$. With $J$ large, this difference is strongly negative even for small $\delta$ — meaning the log-likelihood drops sharply as $\theta$ moves away from $\hat\theta$. The peak is sharp.

**Step 4 — By (1) and Cramér–Rao:** The score variance $F(\theta)$ measures how informatively each observation speaks about $\theta$. By Cramér–Rao, $\mathrm{Var}(\hat\theta) \geq 1/[n F(\theta)]$, so large $F$ implies $\hat\theta$ has low variance — it is precisely determined by the data.

**Conclusion.** A sensitive distribution (small $\delta$ moves $p_\theta$ a lot in KL) means $F$ is large, the log-likelihood is sharply peaked, the data strongly discriminates between nearby parameter values, and the MLE is well identified — i.e., the data carries a lot of information about $\theta$.

## Reverse Direction: Insensitive Distribution ⇒ Uninformative Data

**Premise.** A small change $\delta$ in $\theta$ produces an almost-identical distribution — formally, $D_{KL}\!\left(p_\theta \,\|\, p_{\theta + \delta}\right)$ is small even for non-trivial $\|\delta\|$.

**Step 1 — By (4):** Since $D_{KL} \approx \tfrac{1}{2}\, \delta^\top F(\theta)\, \delta$ is small, the matrix $F(\theta)$ must be small.

**Step 2 — By (3):** Consequently $J(\theta) \approx n\, F(\theta)$ is also small.

**Step 3 — By (2):** With $J$ small, $\ell(\hat\theta + \delta) - \ell(\hat\theta) \approx -\tfrac{1}{2}\, \delta^\top J(\hat\theta)\, \delta$ is barely negative even for substantial $\delta$. The log-likelihood is broad and flat near $\hat\theta$ — many $\theta$ values fit the data nearly equally well.

**Step 4 — By (1) and Cramér–Rao:** Small $F$ means the score has low variance — each observation provides weak evidence about $\theta$. The Cramér–Rao bound $\mathrm{Var}(\hat\theta) \geq 1/[n F(\theta)]$ now gives a *large* lower bound, so $\hat\theta$ has **high variance** — it is poorly determined by the data.

**Conclusion.** An insensitive distribution (small $\delta$ barely moves $p_\theta$ in KL) means $F$ is small, the log-likelihood is flat, the data cannot discriminate between nearby parameter values, and the MLE is poorly identified — i.e., the data carries little information about $\theta$.

## The Unified Statement

The four notions are mutually equivalent:

$$\underbrace{\text{Sensitivity of } p_\theta \text{ to } \theta}_{D_{KL} \text{ grows fast with } \delta} \;\;\Longleftrightarrow\;\; \underbrace{F(\theta) \text{ large}}_{\text{score variance, KL curvature}} \;\;\Longleftrightarrow\;\; \underbrace{\text{Sharp likelihood peak}}_{J(\hat\theta) \text{ large}} \;\;\Longleftrightarrow\;\; \underbrace{\hat\theta \text{ well identified}}_{\mathrm{Var}(\hat\theta) \text{ small}}$$

and symmetrically, the negations are mutually equivalent:

$$\underbrace{\text{Insensitivity of } p_\theta \text{ to } \theta}_{D_{KL} \text{ small for moderate } \delta} \;\;\Longleftrightarrow\;\; \underbrace{F(\theta) \text{ small}}_{\text{weak score, flat KL bowl}} \;\;\Longleftrightarrow\;\; \underbrace{\text{Flat likelihood peak}}_{J(\hat\theta) \text{ small}} \;\;\Longleftrightarrow\;\; \underbrace{\hat\theta \text{ poorly identified}}_{\mathrm{Var}(\hat\theta) \text{ large}}$$
