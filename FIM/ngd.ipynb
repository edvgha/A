## Vanilla gradient descent, recast

Vanilla GD can be written as the solution to a constrained optimization problem at each step. At parameter $\theta$, we ask:

$$\delta^* = \arg\min_{\delta} \big[L(\theta) + \nabla L(\theta)^\top \delta\big] \quad \text{subject to} \quad \|\delta\|_2^2 \leq \epsilon^2$$

In words: "Find the $\delta$ that decreases $L$ the most, among all $\delta$'s of bounded Euclidean length."

Solving this (via Lagrange multipliers) gives:

$$\delta^* \propto -\nabla L(\theta).$$

So vanilla GD is steepest descent with respect to Euclidean distance in parameter space.

## Natural gradient descent, derived

Now replace the Euclidean constraint with the statistically meaningful one:

$$\delta^* = \arg\min_{\delta} \big[L(\theta) + \nabla L(\theta)^\top \delta\big] \quad \text{subject to} \quad \tfrac{1}{2}\delta^\top F(\theta)\, \delta \leq \epsilon^2$$

In words: "Find the $\delta$ that decreases $L$ the most, among all $\delta$'s that change the distribution by at most $\epsilon$ in KL."

This is identical to vanilla GD except we've swapped Euclidean length for distributional length. The constraint comes directly from the key statement: $\tfrac{1}{2}\delta^\top F \delta$ is (the leading-order term of) the KL divergence.

Solving via Lagrange multipliers:

$$\mathcal{L}(\delta, \lambda) = \nabla L^\top \delta + \lambda\!\left(\tfrac{1}{2} \delta^\top F\, \delta - \epsilon^2\right)$$

$$\frac{\partial \mathcal{L}}{\partial \delta} = \nabla L + \lambda F\, \delta = 0 \;\;\Longrightarrow\;\; \delta = -\tfrac{1}{\lambda}\, F^{-1} \nabla L$$

Absorbing $1/\lambda$ into a learning rate $\eta$:

$$\boxed{\;\delta_{\text{NGD}} = -\eta\, F(\theta)^{-1} \nabla L(\theta)\;}$$

This is natural gradient descent. And it was derived directly from the statement we started with — the moment you accept that $\delta^\top F \delta$ measures real distributional change, NGD is forced upon you.

#### Algorithm

```
Input: initial θ, learning rate η, data D
For t = 1, 2, ...:
    1. Sample mini-batch B from D
    2. Compute gradient:     g  = ∇_θ L(θ)            on B
    3. Estimate FIM:         F̂  = (1/|B|) Σ s(x_i)s(x_i)ᵀ   
                                  where s(x_i) = ∇_θ log p(x_i|θ)
    4. Solve linear system:  d = F̂⁻¹ g            (or use conjugate gradient)
    5. Update parameters:    θ ← θ − η · d
Return θ
```
