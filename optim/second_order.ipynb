#### Newton's step

$$\theta_{t+1} = \theta_t - \eta_t H_t^{-1} g_t$$

#### Optimality condition

#### Necessary

If $\theta^*$ is a local minimum, then it must satisfy $g^* = 0$ and $H^* \succeq 0$ (Positive Semidefinite)

#### Sufficient

If $g^* = 0$ and $H^* \succ 0$ (Positive Definite), then $\theta^*$ is a local minimum

#### Reasoning

$$f(\theta^* + \Delta \theta) \approx f(\theta^*) + (g^*)^\top \Delta \theta + \frac{1}{2} (\Delta \theta)^\top H^* \Delta \theta$$

$$\Delta f \approx \frac{1}{2} (\Delta \theta)^\top H^* \Delta \theta$$

$\Delta f \geq 0$ - No direction to decrease

#### Eigenvalue connection

$$\Delta f \approx g^\top \Delta \theta + \frac{1}{2} \Delta \theta^\top H \Delta \theta$$

$$\Delta \theta = -H^{-1}g$$

replace $\Delta \theta$ in approximation

$$\Delta f \approx -\frac{1}{2} g^\top H^{-1} g$$

$$H^{-1} = (V \Lambda V^\top)^{-1} = V \Lambda^{-1} V^\top$$
$$H^{-1} = \sum_{i=1}^{n} \frac{1}{\lambda_i} v_i v_i^\top$$

$$\Delta f \approx -\frac{1}{2} \sum_{i=1}^{n} \frac{(v_i^\top g)^2}{\lambda_i}$$

The term $(v_i^\top g)^2$ is always non-negative because it is a squared scalar. Therefore, the sign of each term in the summation is determined solely by the eigenvalue $\lambda_i$ and the leading negative sign:Positive Eigenvalues ($\lambda_i > 0$): The term $-\frac{1}{2} \frac{(v_i^\top g)^2}{\lambda_i}$ is negative, meaning the update decreases the function value in these directions.Negative Eigenvalues ($\lambda_i < 0$): The term $-\frac{1}{2} \frac{(v_i^\top g)^2}{\lambda_i}$ becomes positive (a negative divided by a negative). This means the update actually increases the function value along the directions of negative curvature.Summary of the Failure MechanismIf the gradient $g$ has a significant component aligned with eigenvectors corresponding to negative eigenvalues, the sum will be dominated by positive terms. In this scenario, $\Delta f > 0$, and the Newton step moves "uphill" toward a maximum or saddle point instead of "downhill" toward a minimum.

### Saddle-Free Newton 

Instead of using the raw Hessian $H$, SFN uses the absolute value of the Hessian, denoted as $|H|$. The update becomes:

$$\Delta \theta = -|H|^{-1} g$$

$$\Delta \theta = -\sum_{i=1}^{n} \frac{1}{|\lambda_i|} (v_i^\top g) v_i$$

$$\Delta \theta = -(V |\Lambda| V^\top)^{-1} g$$

### Hessian damping

$$\Delta \theta = -(H + \lambda I)^{-1} g$$

By adding $\lambda I$ (where $\lambda > 0$ is the damping parameter and $I$ is the identity matrix), you are effectively shifting the entire eigenvalue spectrum of the Hessian:

* Spectral Shift: If the original eigenvalues of $H$ are $\{\lambda_1, \lambda_2, \dots, \lambda_n\}$, the eigenvalues of the damped Hessian $(H + \lambda I)$ become $\{\lambda_1 + \lambda, \lambda_2 + \lambda, \dots, \lambda_n + \lambda\}$.

* Ensuring Positivity: Even if $H$ has negative or zero eigenvalues, a sufficiently large $\lambda$ ensures that all eigenvalues of the modified matrix are positive, making it positive definite and invertible.

Hessian damping is mathematically equivalent to the Trust Region method

In a Trust Region method, you solve a constrained optimization problem: minimize the Taylor expansion subject to $||\Delta \theta|| \leq \delta$

$$\text{minimize } g^\top \Delta \theta + \frac{1}{2} \Delta \theta^\top H \Delta \theta + \frac{\lambda}{2} ||\Delta \theta||^2$$

Resulting Update: Setting the derivative to zero yields the damped Newton step:

$$(H + \lambda I)\Delta \theta = -g$$

$$\Delta \theta = -(H + \lambda I)^{-1} g$$
