## 1. The model

Each feature $i$ is binary: $x_i \in \{0, 1\}$ (word present or absent). Per class $c$ and feature $i$, a single parameter $\phi_{ci} = P(x_i = 1 \mid c)$. The likelihood of one feature value is the Bernoulli pmf:

$$P(x_i \mid c) = \phi_{ci}^{\,x_i}\,(1-\phi_{ci})^{\,1-x_i}$$

## 2. The likelihood over a class

Let class $c$ have $N_c$ samples, and let $x_i^{(s)}$ be feature $i$ of sample $s$. By the naive independence assumption across features and i.i.d. across samples:

$$P(\text{data} \mid \phi_c) = \prod_{s=1}^{N_c} \prod_i \phi_{ci}^{\,x_i^{(s)}} (1-\phi_{ci})^{\,1-x_i^{(s)}}$$

Focus on one feature $i$ (they separate). Define $N_{ci} = \sum_s x_i^{(s)}$ = number of class-$c$ samples where feature $i$ is present. Then collecting the exponents across the $N_c$ samples:

$$P_i(\text{data} \mid \phi_{ci}) = \phi_{ci}^{\,N_{ci}}\,(1-\phi_{ci})^{\,N_c - N_{ci}}$$

(the present count is $N_{ci}$; the absent count is $N_c - N_{ci}$).

## 3. The prior — a Beta (two-outcome Dirichlet)

A Bernoulli parameter $\phi_{ci} \in [0,1]$ has a natural conjugate prior: the Beta distribution, which is the Dirichlet specialized to two outcomes (present/absent). The symmetric Beta with concentration $\alpha+1$ on both outcomes:

$$P(\phi_{ci}) \propto \phi_{ci}^{\,\alpha}\,(1-\phi_{ci})^{\,\alpha}$$

## 4. MAP estimate — maximize the log-posterior

Posterior $\propto$ likelihood $\times$ prior; add exponents on the matching bases:

$$P(\phi_{ci} \mid \text{data}) \propto \phi_{ci}^{\,N_{ci}}(1-\phi_{ci})^{\,N_c-N_{ci}} \cdot \phi_{ci}^{\,\alpha}(1-\phi_{ci})^{\,\alpha} = \phi_{ci}^{\,N_{ci}+\alpha}\,(1-\phi_{ci})^{\,N_c-N_{ci}+\alpha}$$

Take the log:

$$\log P = (N_{ci}+\alpha)\log\phi_{ci} + (N_c - N_{ci} + \alpha)\log(1-\phi_{ci}) + \text{const}$$

Differentiate w.r.t. $\phi_{ci}$ and set to zero (no Lagrange needed — $\phi$ and $1-\phi$ handle the constraint automatically):

$$\frac{d\log P}{d\phi_{ci}} = \frac{N_{ci}+\alpha}{\phi_{ci}} - \frac{N_c - N_{ci} + \alpha}{1-\phi_{ci}} = 0$$

Cross-multiply:

$$(N_{ci}+\alpha)(1-\phi_{ci}) = (N_c - N_{ci}+\alpha)\,\phi_{ci}$$

Expand the left side and collect $\phi_{ci}$:

$$(N_{ci}+\alpha) - (N_{ci}+\alpha)\phi_{ci} = (N_c - N_{ci}+\alpha)\phi_{ci}$$

$$(N_{ci}+\alpha) = \phi_{ci}\big[(N_{ci}+\alpha) + (N_c - N_{ci}+\alpha)\big] = \phi_{ci}\big[N_c + 2\alpha\big]$$

The two $\alpha$'s sum and the $\pm N_{ci}$ cancel, leaving $N_c + 2\alpha$. Therefore:

$$\boxed{P(x_i = 1 \mid c) = \frac{N_{ci} + \alpha}{N_c + 2\alpha}}$$

## 5. Scoring

Unlike multinomial, Bernoulli scores absent features too — the $(1-\phi)$ term contributes when $x_i = 0$:

$$\log P(c \mid x) \;\overset{+}{=}\; \log P(c) + \sum_i \Big[ x_i \log\phi_{ci} + (1-x_i)\log(1-\phi_{ci}) \Big]$$
