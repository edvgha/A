## 1. The model

Each feature $i$ is continuous. Per class $c$ and feature $i$, model it as normal with parameters $\mu_{ci}, \sigma_{ci}^2$:

$$P(x_i \mid c) = \frac{1}{\sqrt{2\pi\sigma_{ci}^2}}\exp\!\left(-\frac{(x_i-\mu_{ci})^2}{2\sigma_{ci}^2}\right)$$

No conjugate-prior smoothing here â€” we do plain MLE and floor the variance afterward.

## 2. The likelihood

Over the $N_c$ samples in class $c$, for feature $i$ (with values $x_i^{(s)}$):

$$P_i(\text{data} \mid \mu_{ci}, \sigma_{ci}^2) = \prod_{s=1}^{N_c} \frac{1}{\sqrt{2\pi\sigma_{ci}^2}}\exp\!\left(-\frac{(x_i^{(s)}-\mu_{ci})^2}{2\sigma_{ci}^2}\right)$$

Take the log (turns the product into a sum, and the exp vanishes):

$$\ell = -\frac{N_c}{2}\log(2\pi\sigma_{ci}^2) - \frac{1}{2\sigma_{ci}^2}\sum_{s=1}^{N_c}\big(x_i^{(s)}-\mu_{ci}\big)^2$$

## 3. MLE for the mean $\mu_{ci}$

Differentiate $\ell$ w.r.t. $\mu_{ci}$. Only the second term depends on it; chain rule on the square:

$$\frac{\partial \ell}{\partial \mu_{ci}} = -\frac{1}{2\sigma_{ci}^2}\sum_s 2\big(x_i^{(s)}-\mu_{ci}\big)(-1) = \frac{1}{\sigma_{ci}^2}\sum_s \big(x_i^{(s)}-\mu_{ci}\big) = 0$$

The factor $1/\sigma_{ci}^2 > 0$ divides out, leaving $\sum_s (x_i^{(s)} - \mu_{ci}) = 0$, i.e. $\sum_s x_i^{(s)} = N_c\,\mu_{ci}$:

$$\boxed{\mu_{ci} = \frac{1}{N_c}\sum_{s=1}^{N_c} x_i^{(s)}}$$

## 4. MLE for the variance $\sigma_{ci}^2$

Let $v = \sigma_{ci}^2$ and differentiate $\ell$ w.r.t. $v$:

$$\ell = -\frac{N_c}{2}\log(2\pi v) - \frac{1}{2v}\sum_s (x_i^{(s)}-\mu_{ci})^2$$

$$\frac{\partial \ell}{\partial v} = -\frac{N_c}{2v} + \frac{1}{2v^2}\sum_s (x_i^{(s)}-\mu_{ci})^2 = 0$$

Multiply through by $2v^2$:

$$-N_c\,v + \sum_s (x_i^{(s)}-\mu_{ci})^2 = 0$$

$$\boxed{\sigma_{ci}^2 = \frac{1}{N_c}\sum_{s=1}^{N_c}\big(x_i^{(s)}-\mu_{ci}\big)^2}$$

The MLE variance is the sample variance (the biased $1/N_c$ version, not $1/(N_c-1)$ â€” MLE always gives the biased one). In practice add a small floor $\sigma_{ci}^2 \leftarrow \sigma_{ci}^2 + \epsilon$ to prevent division by zero when a feature is constant within a class â€” the Gaussian analogue of smoothing.

## 5. Scoring

Plug the fitted $\mu_{ci}, \sigma_{ci}^2$ into the log-density and sum over features:

$$\log P(c \mid x) \;\overset{+}{=}\; \log P(c) + \sum_i \left[ -\frac{1}{2}\log(2\pi\sigma_{ci}^2) - \frac{(x_i-\mu_{ci})^2}{2\sigma_{ci}^2} \right]$$
